```python
import numpy as np

#Función ackley 1
def ackley_1(x):
    D = 10
    sum1 = 0.0
    sum2= 0.0
    for g in x :
        sum1 = sum1 + g**2.0
        sum2 = sum2 + np.cos(2.0*np.pi*g)
    return -20.0*np.exp(-0.02*np.sqrt(sum1/D))-np.exp(sum2/D) + 20 + np.e

#Función ackley 2
def ackley_2(x1, x2):
    term1 = -200*np.exp(-0.02*np.sqrt(x1**2 + x2**2))
    return term1

#Función ackley 3
def ackley_3(x, y):
    return -200*np.exp(-0.02*np.sqrt(x**2 + y**2)) + 5*np.exp(np.cos(3*x)+np.sin(3*y))

#Función ackley 4
def ackley_4(x):
    b = 0.2
    resultado = 0
    for i in range(0, len(x)-1, 2):
        numero=x[i]
        numero2=x[i+1]
        resultado = resultado + (np.exp(-b) * np.sqrt((numero**2) + (numero2**2)) + 3 * (np.cos(2*numero) + np.sin(2*numero2)))
    return resultado



#Algoritmo de evolución diferencial
def evolucionDif(funcionFitness, limites, tam_poblacion,max_gen, F=0.5, Cr=0.6):
    D = len(limites)
    
    # Generación de la población inicial
    population = np.random.uniform(limites[:, 0], limites[:, 1],size=(tam_poblacion, D))
    
    #Inicializar con ceros el vector de valores fitness
    fitness = np.zeros(tam_poblacion)
    
    #Primer llenado del vector con la población inicial,esto es para sacar el primer "Mejor valor"
    for i in range(tam_poblacion):
        if funcionFitness.__name__ in ['ackley_1', 'ackley_4']:
            fitness[i] = funcionFitness(population[i])
        else:
            fitness[i] = funcionFitness(*population[i])
    
    best_fitness = np.min(fitness)
    best_solution = population[np.argmin(fitness)]
    
    for generation in range(max_gen):
        for i in range(tam_poblacion):
            indices = [indice for indice in range(tam_poblacion) if indice != i]
            
            #Elige 3 valores de indice,sin que sean repetidos.
            a, b, c = np.random.choice(indices, size=3, replace=False)
            #Mutacion
            mutant = population[a] + F * (population[b] - population[c])
            mutant = np.clip(mutant, limites[:, 0], limites[:, 1])
            trial = np.copy(population[i])
            j_rand = np.random.randint(D)
            for j in range(D):
                if np.random.rand() < Cr or j == j_rand:
                    trial[j] = mutant[j]
            if funcionFitness.__name__ in ['ackley_1', 'ackley_4']:
                trial_fitness = funcionFitness(trial)
            else:
                trial_fitness = funcionFitness(*trial)        
            if trial_fitness < fitness[i]:
                population[i] = trial
                fitness[i] = trial_fitness
                if trial_fitness < best_fitness:
                    best_fitness = trial_fitness
                    best_solution = trial
    return best_solution, best_fitness

# Parámetros
tam_poblacion = 50
max_gen = 5000
F = 0.5
Cr = 0.6
D = 10 #Lo dice la consigna del enunciado

#Asignación de limites
limites_1 = np.array([(-35, 35)] * D) 
limites_2 = np.array([(-32, 32), (-32, 32)])
limites_3 = np.array([(-32, 32), (-32, 32)])
limites_4 = np.array([(-35, 35)] * D)

resultados = []

for run in range(25):
    np.random.seed(run) 
    best_solution_1, best_fitness_1 = evolucionDif(ackley_1, limites_1,tam_poblacion, max_gen, F, Cr)
    best_solution_2, best_fitness_2 = evolucionDif(ackley_2, limites_2,tam_poblacion, max_gen, F, Cr)
    best_solution_3, best_fitness_3 = evolucionDif(ackley_3, limites_3,tam_poblacion, max_gen, F, Cr)
    best_solution_4, best_fitness_4 = evolucionDif(ackley_4, limites_4,tam_poblacion, max_gen, F, Cr)
    resultados.append({
        "Ackley 1": (best_solution_1, best_fitness_1),
        "Ackley 2": (best_solution_2, best_fitness_2),
        "Ackley 3": (best_solution_3, best_fitness_3),
        "Ackley 4": (best_solution_4, best_fitness_4)
    })

# Mostrar los resultados de cada ejecución
for run, result in enumerate(resultados):
    print(f"Ejecución {run+1}:")
    print("Mejor solución encontrada para Ackley 1:")
    print(" Vector de entrada:", result["Ackley 1"][0])
    print(" Valor de la función en la mejor solución:", result["Ackley 1"][1])
    print()
    print("Mejor solución encontrada para Ackley 2:")
    print(" x1 =", result["Ackley 2"][0][0])
    print(" x2 =", result["Ackley 2"][0][1])
    print(" Valor de la función en la mejor solución:", result["Ackley 2"][1])
    print()
    print("Mejor solución encontrada para Ackley 3:")
    print(" x1 =", result["Ackley 3"][0][0])
    print(" x2 =", result["Ackley 3"][0][1])
    print(" Valor de la función en la mejor solución:", result["Ackley 3"][1])
    print()
    print("Mejor solución encontrada para Ackley 4:")
    print(" Solución =", result["Ackley 4"][0])
    print(" Valor de la función en la mejor solución:", result["Ackley 4"][1])
    print()
    
# Calcular los mejores valores objetivo, mediana y desviación
fitness_values_1 = [result["Ackley 1"][1] for result in resultados]
fitness_values_2 = [result["Ackley 2"][1] for result in resultados]
fitness_values_3 = [result["Ackley 3"][1] for result in resultados]
fitness_values_4 = [result["Ackley 4"][1] for result in resultados]

best_fitness_1 = np.min(fitness_values_1)
best_fitness_2 = np.min(fitness_values_2)
best_fitness_3 = np.min(fitness_values_3)
best_fitness_4 = np.min(fitness_values_4)

median_fitness_1 = np.median(fitness_values_1)
median_fitness_2 = np.median(fitness_values_2)
median_fitness_3 = np.median(fitness_values_3)
median_fitness_4 = np.median(fitness_values_4)

std_deviation_1 = np.std(fitness_values_1)
std_deviation_2 = np.std(fitness_values_2)
std_deviation_3 = np.std(fitness_values_3)
std_deviation_4 = np.std(fitness_values_4)

# Mostrar los mejores valores objetivo, mediana y desviación
print("Mejores valores objetivo:")
print(" Ackley 1:", best_fitness_1)
print(" Ackley 2:", best_fitness_2)
print(" Ackley 3:", best_fitness_3)
print(" Ackley 4:", best_fitness_4)
print()
print("Mediana de los valores objetivo:")
print(" Ackley 1:", median_fitness_1)
print(" Ackley 2:", median_fitness_2)
print(" Ackley 3:", median_fitness_3)
print(" Ackley 4:", median_fitness_4)
print()
print("Desviación estándar de los valores objetivo:")
print(" Ackley 1:", std_deviation_1)
print(" Ackley 2:", std_deviation_2)
print(" Ackley 3:", std_deviation_3)
print(" Ackley 4:", std_deviation_4)
```
